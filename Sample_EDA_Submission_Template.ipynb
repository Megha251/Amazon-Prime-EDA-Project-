## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
data=titles.merge(credits,on='id')
````

````python
data.head()
````

````python
data=data.drop_duplicates()
````

````python
data=data.dropna()
````

````python
data=data[['id','title','type','release_year','genres','imdb_score','name','role']]
````

### What all manipulations have you done and insights you found?

In this project, several data wrangling steps were performed to prepare the dataset for analysis. First, the two datasets (titles and credits) were merged using a common column 'id'. This helped in combining movie details with cast and crew information.

Next, duplicate rows were removed to ensure data quality. Missing values were handled by removing rows with null values to make the dataset clean and consistent.

Only relevant columns were selected for better analysis, such as title, type, release year, genres, IMDb score, and cast details.

From the analysis, it was observed that the dataset contains both movies and TV shows with different genres and ratings. Some columns had missing values which were cleaned during preprocessing. The data can now be used to analyze trends such as popular genres, ratings distribution, and content type comparison on Amazon Prime.
